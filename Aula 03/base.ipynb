```python
def geradorDeDesconto(valor, desconto = 0.1):
    """
        Função calcula um desconto por % inserido.
        valor: Pode ser float ou int
        desconto: Precisa ser informado em decimal
    """

    return valor * desconto

geradorDeDesconto(100, 0.2)
```

```python
import json

try:
    with open(r"dados\contas.json", 'r', encoding="UTF8") as arquivo:
        contas = json.load(arquivo)

except FileNotFoundError as e:
    print(e)
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/>
        Tratamento
    </strong>
</p>

```python
# Converter data em texto para data real (datetime)

from datetime import datetime

for conta in contas:
    conta['vencimento'] = datetime.strptime(conta['vencimento'], "%d/%m/%Y")                              
```

```python
# Verificando conversão
contas[0]['vencimento']
```

<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong><br/><br/><br/><br/>
        Análise (Vencimentos)
    </strong>
</p>

```python
# Quantas contas estão vencidas?

from datetime import datetime

hoje = datetime.today()
contas_vencidas = []

for conta in contas:
    if conta['vencimento'] < hoje and conta['status'] == "Em aberto" and conta['tipo'] == "Pagar":
        contas_vencidas.append(conta)

len(contas_vencidas)
```

```python
# Total em contas vencidas?

total_vencido = 0

for conta in contas_vencidas:
    total_vencido += conta['valor']

round(total_vencido, 2)
```
