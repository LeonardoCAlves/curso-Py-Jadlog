<p style="font-size: 1.1rem; color:#9beb3b; text-align: center; font-family: monospace ">
    <strong></br></br></br></br>
        Validando um senha de brincadeira
    </strong>
</p>

```python
senha_cadastrada = input('Senha: ')
senha_entrada = ''
tentativas = 0

while senha_cadastrada != senha_entrada:
    senha_entrada = input('Digite sua senha: ')
    tentativas += 1

    if tentativas == 3 and senha_cadastrada != senha_entrada:
        print("Tentativas máximas atingidas")
        break

if senha_cadastrada == senha_entrada:
    print('Acesso permitido')
```

```python
while True:
    menu = int(input('1_Cadastrar | 2.Visualizar | 3.Sair | Opção: '))

    match menu:
        case 1:
            nome_franquia = input('Franquia: ')
        case 2:
            print(nome_franquia)
        case 3:
            break
        case _:
            print('Opção inválida!')
            
   
```

        # Exemplo: Imagine que cada franquiado me pague 10% de seu faturamento mensal 
    </strong>
</p>

```python
# Entrada de valores

franquias = ['RJ Rio de Janeiro 55', 'LJ São Paulo 55', 'LJ Três Rios 11']

pg_franquias = []

for franquia in franquias:
    valor = float(input(f'{franquia} | Valor: '))

    pg_franquias.append(franquia)
    pg_franquias.append(valor)    

pg_franquias
```

```python
for i, d in enumerate(pg_franquias):
    if i % 2 == 1:
        print(d)
```

```python
# Cálculo
soma = 0
for indice, dado in enumerate(pg_franquias):
    if indice % 2 == 1:
        # soma = soma + dado * 0.1
        soma += dado * 0.1
        
print(soma)
```

        # Exemplo: Cadastrando e consumindo Franquias  
    </strong>
</p>

```python
franquias_cadastradas = []

for contator in range(2):
 
    franquia = dict(
        id       = int(input('Identificação: ')),
        cpf      = input('CPF: '),
        endereco = input('Endereço: ').title(),
        bairro   = input('Bairro: ').title(),
        cidade   = input('Cidade: ').title(),
        estado   = input('Estado: ').upper()
    )
    franquias_cadastradas.append(franquia)
```

```python
for indice, franquia in enumerate(franquias_cadastradas):     # Lista
    print(f"{indice + 1}ª Franquia\n")

    for chave, valor in franquia.items():                     #Dicionário
        print(f'{chave:<10} → {valor}')
    print()
```
