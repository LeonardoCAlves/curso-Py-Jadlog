```python
saldo = 0
lista_transacoes = []

while True:
    menu = int(input('1.Dep |2.Saq |3.Sair | Op√ß√£o: '))
    
    if menu in (1, 2):
        valor = float(input('Valor: '))

    match menu:
        case 1:
            print(f'Dep√≥sito conclu√≠do. | R$ {valor} ')
            # saldo = saldo + valor
            saldo += valor
            lista_transacoes.append(valor)

        case 2:
            if valor <= saldo:
                saldo -= valor
                print(f'Saque conclu√≠do. | Saldo atual: R$ {saldo}')
                lista_transacoes.append(valor * -1)

            else:
                print(f'Saldo insuficiente! | Saldo atual: R$ {saldo}')
        case 3:
            break
        case _:
            print('Op√ß√£o inv√°lida...')

print(lista_transacoes)
print(saldo)

```
